```python
from sympy import * 
x = symbols('x') 
eq = x**4 - 1 
solutions = solve(eq, x) 
print(solutions) 
```

```python
from sympy import *                   # (i)
num_decimals = 100 
sqrt_2 = N(sqrt(2), num_decimals) 
print(sqrt_2)
```

```python
from sympy import *                           #ii
expr = Rational(1, 2) + Rational(1, 3) 
simplified_expr = simplify(expr) 
print(simplified_expr) 
```

```python
from sympy import *                           #iii
x, y = symbols('x y') 
expr = (x + y)**6 
expanded_expr = expand(expr) 
print(expanded_expr) 
```
